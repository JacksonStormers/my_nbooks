```python
import pandas as pd
import seaborn as sns
```

```python
avocado_data = pd.read_csv('avocado.csv')
avocado_data.head()
```

```python
avocado_data.info(memory_usage='deep')
```

```python
avocado_data['Date'] = pd.to_datetime(avocado_data['Date'])
avocado_data.info(memory_usage='deep')
```

```python
avocado_query = avocado_data.query('year == 2015 and type == "conventional" and region == "Albany"')[['Date', 'Total Bags', 'Small Bags']]
avocado_query.head()
```

```python
avocado_melt = pd.melt(avocado_data, id_vars='Date', value_vars=['Total Bags','Small Bags'], var_name='Bags', value_name='Count')
avocado_melt
```

```python
sns.relplot(data=avocado_melt, kind='line', x='Date', y='Count', hue='Bags')
```

```python
avocado_data['Sales Volume'] = pd.qcut(avocado_data['Total Volume'], q=4, labels=['Poor','Modest','Good','Excellent'])
avocado_data.head()
```

```python
sns.countplot(data=avocado_data, x='year', hue='Sales Volume')
```

```python
melted_bags = pd.melt(avocado_data, id_vars='year', value_vars=['Small Bags','Large Bags','XLarge Bags'], var_name='Bags', value_name='Count')
melted_bags
```

```python
sns.relplot(data=melted_bags, kind='line', x='year', y='Count', hue='Bags')
```
